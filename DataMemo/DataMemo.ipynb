<p><b>Variable of interest:</b> The outcome of a direct indexing lot-level decision of whether to harvest a loss <-> cause tracking error (or slight deviation from the index which is essentially variance so has an EV that doesnt meaningfully impact indicie beta equivalence) for generated tax alpha from harvesting a loss on a given lot (EV positive so a direct gain or alpha) since lots can always be purchased back after 30 days or replaced with a similar correlated asset by identifying multicolinearity.

<b>Classification: </b>The outcome is hence a binary variable or boolean y∈{0, 1} || y∈{false, true} which captures the state of whether to harvest or not at a given time t (fixed time interval, a theoretical dt implementation would probably require too much stochastic PDE modeling though probably something I will try to impliment if I have the time to self study stochastic processes)

This will be a classification problem as the outcome is fundamentally discrete. </p>

<b> Informative Predictors: </b> Unrealized loss (% of lot basis) (biggest factor as it directly correlates to the amount of potential tax alpha), holding period in days (atttributes to # of TLH operatiosn possible in a given tax year, allows for implementation of basic wash sale rules that the model can learn), net realized gains YTD (to determine whether a loss offsets anything), estimated tracking error (or the direct result/cost of harvesting which as described earlier is slight deviation from the benchmark index).

<b>Model goal: </b> predictive-dominant mostly supervised model, with the goals at heart making a decision in a direct indexing portfolio that'll increase total alpha (via tax alpha) in portfolio's that face signficiant amounts of capital gains tax while maintaining indicie beta very simlar if not equal to the underlying index, which involves predicting the best potential decisionand within a given boundary / restriction though that is simlar to the conditions similar to normal direct indexing(an oracle machine of sorts that the ML model roughly follows) of a mechanical direct indexing model (which is generally a conditional statement of >= % loss, harvest). This does unfortunately restrict capabilities of an ML based implementation since they are both using the same underlying condition but with the mechanichal direct indexing model its behavior is restricted on the boundary itself while the ML model's behavior would be defined within the domain that the boundary restricts (its mildly analogous to the generalized stoke's theorem when analyzing manifolds except that the interior information is not conserved to the boundary information). The reason why the ML model has the goal of approximating the oracle machine instead of having free reign on predictions / or an answer key of the right decision is because true harvest labels require data far into the future which would not be possible in actual deployment and even if it was would likely require time series elements. The restriction here with this analogy is since the ML model can only work within the domain that the boundary restricts (oracle conditions) the ML model therefore cannot behave outside that boundary nor can it tell whether the pre-conditions are optimal or suboptimal.

secondary compoennt: inferential + an unsupervised component of dimensionality reduction (likely PCA or factor based reduction) in order to select a proper subset of stocks that can still effectively model the underlying index due to market indicies having signficiant amounts of multicolinearity due to shared factor exposures, sectors, weighted differences, ect. Dimensionality reduction is important in the context of this project for reducing the probability of having too many features and reduces tracking error variance from multicolinear noise. This is the first step before the dominant goal of classification starts.

The model will also have an infernetial goal as åt the high level there is a strong correlation between ML's empirical /flexible elements resulting in less understanding of the why the model is making given predictions comapred to base mechanistic/mathematically smooth, continuous, and differentiable models baked with assumptions that can't account for complexities of modeling real scenarios (underfitting / huge bias out of the bias variance tradeoff).
